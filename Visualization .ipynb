There is also a continuous observation of peaks of and dips of IPOs with their significant probability of being under current gain. The statistical analysis along with fundamental analysis also contributed to the changes as you can see in the graph. The reasons being


1. Addition of number of startups: due to a startup boom in last 2 years
2. Political influence: Elections and governmental schemes, the Covid-19 pandemic
3. Number of investors increasing : Due work from home, job losses, and lockdown, the number of small scale and large scale traders increasing
